```python
# Import CNN Libraries

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import (

    Conv2D,
    MaxPooling2D,
    Flatten,
    Dense,
    Dropout

)
```

```python
# Build Custom CNN Model

cnn_model = Sequential([

    Conv2D(
        32,
        (3,3),
        activation='relu',
        input_shape=(128,128,3)
    ),

    MaxPooling2D(2,2),

    Conv2D(
        64,
        (3,3),
        activation='relu'
    ),

    MaxPooling2D(2,2),

    Conv2D(
        128,
        (3,3),
        activation='relu'
    ),

    MaxPooling2D(2,2),

    Flatten(),

    Dense(
        128,
        activation='relu'
    ),

    Dropout(0.5),

    Dense(
        1,
        activation='sigmoid'
    )
])

print("Custom CNN model created successfully!")
```

```python
# Compile CNN Model

cnn_model.compile(

    optimizer='adam',

    loss='binary_crossentropy',

    metrics=['accuracy']
)

print("CNN model compiled successfully!")
```

```python
# CNN Model Summary

cnn_model.summary()
```

```python
# Train Custom CNN Model

cnn_history = cnn_model.fit(

    image_generator.flow(
        X_train_img,
        y_train_img,
        batch_size=16
    ),

    validation_data=(
        X_test_img,
        y_test_img
    ),

    epochs=10,
    callbacks=[early_stopping]
)
```

```python
# Visualize CNN Accuracy Curves

plt.figure(figsize=(8,5))

plt.plot(
    cnn_history.history['accuracy'],
    label='Training Accuracy'
)

plt.plot(
    cnn_history.history['val_accuracy'],
    label='Validation Accuracy'
)

plt.title("CNN Accuracy Curves")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.legend()

plt.show()
```

```python
# Visualize CNN Loss Curves

plt.figure(figsize=(8,5))

plt.plot(
    cnn_history.history['loss'],
    label='Training Loss'
)

plt.plot(
    cnn_history.history['val_loss'],
    label='Validation Loss'
)

plt.title("CNN Loss Curves")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.legend()

plt.show()
```

```python
# Evaluate CNN Model

cnn_loss, cnn_accuracy = cnn_model.evaluate(
    X_test_img,
    y_test_img
)

print("CNN Test Loss:", cnn_loss)
print("CNN Test Accuracy:", cnn_accuracy)
```

```python
# CNN Predictions

cnn_predictions = cnn_model.predict(X_test_img)

cnn_predictions = (
    cnn_predictions > 0.5
).astype(int)
```

```python
# CNN Classification Report

from sklearn.metrics import (
    classification_report,
    confusion_matrix
)

print(
    classification_report(
        y_test_img,
        cnn_predictions
    )
)
```

```python
# CNN Confusion Matrix

cnn_cm = confusion_matrix(
    y_test_img,
    cnn_predictions
)

print(cnn_cm)
```
