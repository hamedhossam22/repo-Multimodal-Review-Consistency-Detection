```python
# Move Inputs to GPU

input_ids = input_ids.to("cuda")

attention_masks = attention_masks.to("cuda")

print("Inputs moved to GPU successfully!")
```

```python
# Generate BERT Predictions for Fusion

small_input_ids = input_ids[:49]

small_attention_masks = attention_masks[:49]

with torch.no_grad():

    outputs = bert_model(
        input_ids=small_input_ids,
        attention_mask=small_attention_masks
    )

bert_probs = torch.argmax(
    outputs.logits,
    dim=1
).cpu().numpy()

print("BERT Predictions Shape:", bert_probs.shape)
```

```python
# Generate CNN Predictions for Fusion

cnn_probs = cnn_model.predict(
    X_test_img
)

cnn_probs = (
    cnn_probs > 0.5
).astype(int)

cnn_probs = cnn_probs.flatten()

print("CNN Predictions Shape:", cnn_probs.shape)
```

```python
print(len(bert_probs))
print(len(cnn_probs))
```

```python
min_len = min(len(bert_probs), len(cnn_probs))

bert_probs = bert_probs[:min_len]
cnn_probs = cnn_probs[:min_len]
```

```python
print(len(bert_probs))
print(len(cnn_probs))
```

```python
# Create Multimodal Fusion Features

fusion_features = np.column_stack((

    bert_probs,
    cnn_probs

))

print("Fusion Features Shape:", fusion_features.shape)

print(fusion_features[:5])
```

```python
# Multimodal Fusion Decision

fusion_predictions = []

for text_pred, image_pred in fusion_features:

    if text_pred == 1 or image_pred == 1:

        fusion_predictions.append(1)

    else:

        fusion_predictions.append(0)

fusion_predictions = np.array(fusion_predictions)

print("Fusion Predictions Shape:", fusion_predictions.shape)

print(fusion_predictions[:10])
```

```python
print(len(y_test_img))
print(len(fusion_predictions))
```

```python
min_len = min(len(y_test_img), len(fusion_predictions))

y_test_img = y_test_img[:min_len]
fusion_predictions = fusion_predictions[:min_len]
```

```python
print(len(y_test_img))
print(len(fusion_predictions))
```

```python
# Multimodal Fusion Evaluation

print(

    classification_report(
        y_test_img,
        fusion_predictions
    )
)
```
